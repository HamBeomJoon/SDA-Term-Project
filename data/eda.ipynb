```python
delet_col=['지역별(1)','지역별(3)','상품군별(2)']
df_raw.drop(delet_col, axis=1, inplace=True)
```

```python
df_raw.head()
```

```python
df =df_raw.melt(id_vars=["지역별(2)",'상품군별(1)','판매유형별(1)'], var_name="기간", value_name="백만원")
df.shape
```

```python
df.head()
```

```python
df['연도']=df["기간"].map(lambda x: int(x.split('.')[0]))
df.head()
```

```python
df["분기"] = df["기간"].map(lambda x: int(x.split('.')[1].split('/')[0]))
df.head()
```

```python
df["백만원"]=df['백만원'].replace("-",np.nan).astype(float)
```
